Relative standard derivation


```python
# Calculate the mean and standard deviation of the dataset
mean = np.mean(data1)
std_dev = np.std(data1)

# Calculate the relative standard deviation (RSD)
rsd = std_dev / mean * 100

print("Mean:", mean)
print("Standard deviation:", std_dev)
print("Relative Standard Deviation (RSD):", rsd)
```

t-test entre les deux semaines 1: sulfemethazole 2: thrimétoprime

```python
t_stat, p_val = ttest_ind(S_sem1, S_sem2)

print("T-statistic:", t_stat)
print("P-value:", p_val)

t_stat, p_val = ttest_ind(T_sem1, T_sem2)

print("T-statistic:", t_stat)
print("P-value:", p_val)
```

```python
# Perform the Wilcoxon signed-rank test
stat, p_val = wilcoxon(S_sem1, S_sem2)

print("Test statistic:", stat)
print("P-value:", p_val)

# Perform the Wilcoxon signed-rank test
stat, p_val = wilcoxon(T_sem1, T_sem2)

print("Test statistic:", stat)
print("P-value:", p_val)
```

Cohen's text

```python
# Calculate the mean and standard deviation of each replica
mean1, std1 = np.mean(S_sem1), np.std(S_sem1, ddof=1)
mean2, std2 = np.mean(S_sem2), np.std(S_sem2, ddof=1)

# Calculate Cohen's d Sulféméthazole
d = (mean1 - mean2) / np.sqrt((std1 ** 2 + std2 ** 2) / 2)

print("Cohen's d:", d)

# Calculate the mean and standard deviation of each replica
mean1, std1 = np.mean(T_sem1), np.std(T_sem1, ddof=1)
mean2, std2 = np.mean(T_sem2), np.std(T_sem2, ddof=1)

# Calculate Cohen's d thrimetoprim
d = (mean1 - mean2) / np.sqrt((std1 ** 2 + std2 ** 2) / 2)

print("Cohen's d:", d)
```
